# 투자 분석 보고서 서비스

현재 디렉토리의 Markdown 문서를 읽어 LangGraph 기반 투자 분석 보고서를 생성합니다. 핵심 Agent 로직은 `agents/` 모듈로 분리되어 있고, 이 노트북은 실행과 점검에 집중합니다.

## 구현 단계

1. 프로젝트 루트와 설정을 확인합니다.
2. `agents.service.InvestmentAnalysisService`를 초기화합니다.
3. 그래프 구조와 정책 분기(`Top3 추천` 또는 `전체 보류`)를 확인합니다.
4. 그래프를 실행하고 결과 보고서를 `outputs/`에 저장합니다.

```python
import sys
from pathlib import Path

NOTEBOOK_DIR = Path.cwd().resolve()
PROJECT_DIR = NOTEBOOK_DIR if (NOTEBOOK_DIR / 'agents').exists() else NOTEBOOK_DIR.parent
if str(PROJECT_DIR) not in sys.path:
    sys.path.insert(0, str(PROJECT_DIR))

from agents import GraphState, InvestmentAnalysisService, ServiceConfig

BASE_DIR = InvestmentAnalysisService.resolve_base_dir(PROJECT_DIR)
BASE_DIR
```

```python
config = ServiceConfig(
    domain="AI Semiconductor",
    recommendation_threshold=20,
    top_k_companies=3,
)
config
```

```python
service = InvestmentAnalysisService(base_dir=BASE_DIR, config=config)
print('source_files:', [path.name for path in service.source_files])
print('chunks:', len(service.split_documents))
```

```python
graph = service.build_graph()
graph
```

```python
initial_state = GraphState(domain=config.domain)
result = graph.invoke(initial_state)
result
```

```python
print('policy_decision:', result.policy_decision)
print('policy_reason:', result.policy_reason)
print('selected_companies:', [item['company_name'] for item in result.selected_companies])
print('hold_companies:', [item['company_name'] for item in result.hold_companies])
print('output_path:', result.output_path)
```
